```python
import copy
class Solution(object):
    def totalNQueens(self, n):
        """
        :type n: int
        :rtype: int
        """
        board = [[0 for _ in range(n)]  for _ in range(n)]
        row = len(board)
        col = len(board[0])
        
        
        def move(row, col, board):
            board_cp = copy.deepcopy(board)
            if board_cp[row][col] == 0:
                board_cp[row][col] = 2

                
                for i in range(len(board[0])):
                    # Already Placed
                    if i == col:
                        continue
                    else:
                        if board_cp[row][i] == 2:
                            return (board,False, row, col)
                        else:
                            board_cp[row][i] = 1
                for i in range(len(board)):
                    if i == row:
                        continue
                    else:
                        if board_cp[i][col] == 2:
                            return (board, False, row, col)
                        else:
                            board_cp[i][col] = 1

                
                invRow = row
                invCol = col
                
                xRow1 = row
                xRow2 = row
                
                xCol1 = col
                xCol2 = col
                # 0
                #  0
                while row + 1 < len(board) and col + 1 < len(board[0]):
                    
                    if board_cp[row+1][col + 1] == 2:
                        return (board, False, row, col)
                    else:
                        board_cp[row+1][col+1] = 1
                        row += 1
                        col += 1
                # 0
                #.  0
                while invRow - 1>= 0 and invCol - 1>= 0:
                    if board_cp[invRow - 1][invCol - 1] == 2:
                        return (board, False, row, col)
                    else:
                        board_cp[invRow - 1][invCol - 1] = 1
                        invRow -= 1
                        invCol -= 1

                while xRow1 + 1 < len(board) and xCol1 - 1 >= 0:
                    if board_cp[xRow1 + 1][xCol1 - 1] == 2:
                        return (board, False, row, col)
                    else:
                        board_cp[xRow1 + 1][xCol1 - 1] = 1
                        xRow1 += 1
                        xCol1 -= 1
                
                while xRow2 - 1 >= 0 and xCol2 + 1 < len(board[0]):
                    if board_cp[xRow2 - 1][xCol2 + 1] == 2:
                        return (board, False, row, col)
                    else:
                        board_cp[xRow2 - 1][xCol2 + 1] = 1
                        xRow2 -= 1
                        xCol2 += 1

#                 print("NEW BOARD")   
#                 showBoard(board_cp)
                return (board_cp, True, row + 1, col)
            else:
                return (board, False, row, col)
        
        result = []
        board_Original = board
        
        for i in range(len(board[0])):
            currRow = 0
            currCol = i
            board = copy.deepcopy(board_Original)
            
            while currRow < row:
                while currCol < col:
                    newBoard, poss, newrow, newcol = move(currRow, currCol, board)
#                     print("i, j : ", currRow, currCol)
#                     showBoard(newBoard)
#                     showBoard("\n")
                    if poss == True:
                        currRow += 1
                        currCol = 0
                        board = newBoard
                        break
                    elif poss == False:
                        currCol += 1                    
                        board = newBoard     
                if currCol == 0:
                    continue
                else:
                    board = []
                    break
            result.append(board)
        result = [x for x in result if len(x) > 0]
        for x in result:
            showBoard(x)
            print("\n")
        return len(result)
        final = []
        for x in result:
            dummy = []
            for row in range(len(x)):
                rowResult = ""
                for col in range(len(x[0])):
                    if x[row][col] == 1:
                        rowResult += "."
                    else:
                        rowResult += "Q"
                dummy.append(rowResult)
            final.append(dummy)
        return final
```

```python
def showBoard(arr):
    for i in range(len(arr)):
        for j in range(len(arr[0])):
            print(arr[i][j], end = " ")
        print("\n")
```
