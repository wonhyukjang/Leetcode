#  Flatten a Multilevel Doubly Linked List



You are given a doubly linked list which in addition to the next and previous pointers, it could have a child pointer, which may or may not point to a separate doubly linked list. These child lists may have one or more children of their own, and so on, to produce a multilevel data structure, as shown in the example below.

Flatten the list so that all the nodes appear in a single-level, doubly linked list. You are given the head of the first level of the list.

 

##### Example 1:



<b>Input:</b> head = [1,2,3,4,5,6,null,null,null,7,8,9,10,null,null,11,12]<br>
<b>Output:</b> [1,2,3,7,8,11,12,9,10,4,5,6]<br>

# Iterative Solution

```python
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, prev, next, child):
        self.val = val
        self.prev = prev
        self.next = next
        self.child = child
"""

from collections import deque
class Solution(object):
    def flatten(self, head):
        """
        :type head: Node
        :rtype: Node
        """
        if head == None:
            return None
        d = []
        d.append(head)
        
        
        result = start = Node(0,None,head,None)
        
        while len(d) > 0:
            tmp = d.pop()

            result.next = tmp
            tmp.prev = result
            result = result.next
            
            if tmp.next:
                d.append(tmp.next)
            
            if tmp.child:
                d.append(tmp.child)
                # In the final result, tmp.child should not exist 
                # We already append child to the stack.
                # Thus, it will handle the rest !
                tmp.child = None 
                
        # our first node.prev should be None !
        start.next.prev = None
                    
        return start.next 
```
