# 46. Permutations


Given a collection of distinct integers, return all possible permutations.



##### Example:



<b>Input:</b> [1,2,3]<br>
<b>Output:</b><br>

[ <br>
  [1,2,3], <br>
  [1,3,2],<br>
  [2,1,3],<br>
  [2,3,1],<br>
  [3,1,2],<br>
  [3,2,1]<br>
]


```python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        result = []
        target = len(nums)
        
        if target == 1:
            return [nums]
        
        def helper(nums):
            if len(nums) == 1:
                return nums
            # Save all possible permutation upto num
            arr = []
            for num in nums:
                # When n = num, add permutation to the result
                if len(nums) == target:
                    dummy = helper([number for number in nums if number != num])
                    
                    # When dummy is just a number
                    if len(dummy) == 1:
                        result.append([num] + dummy)
                        
                    # When dummy is a list
                    else:
                        for x in dummy:
                            result.append([num] + x)
                else:
                    for x in helper([number for number in nums if number != num]):
                        # When num is just a number
                        if type(x) == int:
                            arr.append([num] + [x])
                        # When num is a list                            
                        else:
                            arr.append([num] + x)
            return arr
        helper(nums)
        return result
            
```
