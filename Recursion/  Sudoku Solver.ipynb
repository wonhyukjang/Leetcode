#   Sudoku Solver



```python
import collections


class Solution(object):
    def solveSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: None Do not return anything, modify board in-place instead.
        """
        blank = []
        row = len(board)
        col = len(board[0])
        for i in range(row):
            for j in range(col):
                if board[i][j] == '.':
                    blank.append((i,j))
        
        
        def findMove(blank):
            move = [str(i) for i in range(1,10)]
            row = blank[0]
            col = blank[1]
            
            sectionRow = None
            sectionCol = None
            
            if row >= 0 and row < 3:
                sectionRow = 0
            elif row >= 3 and row < 6:
                sectionRow = 1
            else:
                sectionRow = 2
            
            if col >= 0 and col < 3:
                sectionCol = 0
            elif col >= 3 and col < 6:
                sectionCol = 1
            else:
                sectionCol = 2
            
            section = sectionRow * 3 + sectionCol

#             print(section)
            for i in range(row):            
                if board[i][col] != '.':
                    if board[i][col] in move:
                        move.remove(board[i][col])
            for i in range(col):
                if board[row][i] != '.':
                    if board[row][i] in move:
                        move.remove(board[row][i])

            
            for i in range(3 * sectionRow, 3 * sectionRow + 3):
                for j in range(3 * sectionCol, 3 * sectionCol + 3):
                    if board[i][j] in move:
                        move.remove(board[i][j])
            return move
        
        def showBoard(arr):
            for i in range(len(arr)):
                for j in range(len(arr[0])):                    
                    print(arr[i][j], end = " ")
                    if (j + 1) % 3 == 0:
                        print(" | ", end = "")                                
                print("\n") 
                
        def fill(cell, move):
            row = cell[0]
            col = cell[1]
            board[int(row)][int(col)] = str(move)
            
        def back(cell):
            row = cell[0]
            col = cell[1]
            board[row][col] = '.'
    
            
        def solve(cell ,num):            
            row = blank[num][0]
            col = blank[num][1]
            
            cell = (row,col)
            poss_Move = findMove(cell)

            # If possible Move is none
#             if len(poss_Move) == 0:
#                 return False
            
            if num == len(blank):
                return True

            for move in poss_Move:
                fill(cell, move)
                flag = solve(cell, num + 1)

                if not flag:
                    back(cell)
                else:
                    return True
            return False
        solve(blank[0],0)
```

```python
class Solution:
    def solveSudoku(self, board):
        # Check if num can be placed at board[i][j]
        # O(r + c)
        def can_place(i, j, num):
            # Check row
            if any(num == n for n in board[i]):
                return False
            
            # Check col
            if any(num == n for n in list(zip(*board))[j]):
                return False
            
            # Check box
            start_row, start_col = (i // 3) * 3, (j // 3) * 3
            for r in range(start_row, start_row + 3):
                for c in range(start_col, start_col + 3):
                    if num == board[r][c]:
                        return False
            return True
        
        def showBoard(arr):
            for i in range(len(arr)):
                for j in range(len(arr[0])):                    
                    print(arr[i][j], end = " ")
                    if (j + 1) % 3 == 0:
                        print(" | ", end = "")                                
                print("\n") 

        # O(rc)
        def find_unassigned():
            for r in range(len(board)):
                for c in range(len(board[0])):
                    if board[r][c] == '.':
                        return r, c
            return -1, -1
        
        def solve():
            r, c = find_unassigned()
            
            if r == -1 and c == -1:
                return True
            
            for num in [str(n) for n in range(1, 10)]:
                if can_place(r, c, num):
                    board[r][c] = num
#                     showBoard(board)
                    if solve():
                        return True
                    board[r][c] = '.'
            return False
        if not board:
            return
        solve()
        showBoard(board)
```

```python
class Solution:
    def solveSudoku(self, board):
        rows, cols, triples, visit = collections.defaultdict(set), collections.defaultdict(set), collections.defaultdict(set), collections.deque([])
        for r in range(9):
            for c in range(9):
                if board[r][c] != ".":
                    rows[r].add(board[r][c])
                    cols[c].add(board[r][c])
                    triples[(r // 3, c // 3)].add(board[r][c])
                else:
                    visit.append((r, c))
        def dfs():
            if not visit:
                return True
            r, c = visit[0]
            t = (r // 3, c // 3)
            for dig in {"1", "2", "3", "4", "5", "6", "7", "8", "9"}:
                if dig not in rows[r] and dig not in cols[c] and dig not in triples[t]:
                    board[r][c] = dig
                    rows[r].add(dig)
                    cols[c].add(dig)
                    triples[t].add(dig)
                    visit.popleft()
                    if dfs():
                        return True
                    else:
                        board[r][c] = "."
                        rows[r].discard(dig)
                        cols[c].discard(dig)
                        triples[t].discard(dig)
                        visit.appendleft((r, c))
            return False
        dfs()
```
