#  Construct Binary Tree from Preorder and Inorder Traversal

Given preorder and inorder traversal of a tree, construct the binary tree.

<b>Note:</b><br>
You may assume that duplicates do not exist in the tree.<br>

For example, given<br>
preorder = [3,9,20,15,7]<br>
inorder = [9,3,15,20,7]<br>

Return the following binary tree:

```python
    3
   / \
  9  20
    /  \
   15   7
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: TreeNode
        """
        # preorder 3 9 20 15 7
        # inorder  9 3 15 20 7
        # print(preorder, inorder)
        # if not preorder and not inorder:
        #     return
        if len(preorder) == 0:
            return None
        root = TreeNode(preorder[0])
        # lft = 0
        # for i, elm in enumerate(inorder):
        #     if elm == preorder[0]:
        #         lft = i
        
        # if lft == rt:
        #     return root
        lft = inorder.index(preorder[0])
        root.left = self.buildTree(preorder[1:lft+1], inorder[0:lft])
        # [3], []
        root.right = self.buildTree(preorder[lft+1:], inorder[lft+1:])
        return root
```
